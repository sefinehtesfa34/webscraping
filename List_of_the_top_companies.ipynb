```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
result=requests.get("https://companiesmarketcap.com/")
```

```python
soup=BeautifulSoup(result.text,"lxml")
table=soup.find("table")
header=table.find("tr")
columns=[]
for thead in header.find_all("th"):
  columns.append(thead.text)
df=pd.DataFrame(columns=columns)

for row in table.find_all("tr")[1:]:
  data=row.find_all("td")
  listData=[x.text for x in data]
  df.loc[len(df)]=listData
df.drop(columns="Price (30 days)",inplace=True)
df.Name=df.Name.str[5:]
df["Price"]=df["Price"].str[1:]
df.Country=df.Country.str[3:]
df["Market Cap"]=df["Market Cap"].str[1:-2]
df.Today=df.Today.str[:-1]
listVal=[]
for i in df.Name:
  index=i.index(f"\n")
  if i.lower() not in listVal and  i.upper() not in listVal and  i.capitalize() not in listVal:
    listVal.append(i[:index])
listVal
df.Name=listVal
df.to_excel("top_IT_companies.xlsx",index=False)
read=pd.read_excel("top_IT_companies.xlsx")
```
